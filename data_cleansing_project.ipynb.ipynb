üßπ Step 1: Import Libraries

```python
import pandas as pd
```

üìÇ Step 2: Create Sample Datasets

```python
data1 = {
    'Emp_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', None, 'Eve'],
    'Department': ['HR', 'IT', None, 'Finance', 'IT'],
    'Salary': [50000, 60000, None, 65000, 62000]
}
df1 = pd.DataFrame(data1)
data2 = {
    'Emp_ID': [101, 102, 103, 106],
    'Project': ['Recruitment', 'App Dev', 'Data Analysis', 'Marketing']
}
df2 = pd.DataFrame(data2)
```

üîç Step 3: Check for Missing Data

```python
print(df1.isnull().sum())
```

üß© Step 4: Fill Missing Data

```python
df1['Name'].fillna('Unknown', inplace=True)
df1['Department'].fillna(df1['Department'].mode()[0], inplace=True)
df1['Salary'].fillna(df1['Salary'].mean(), inplace=True)
```

üî§ Step 5: String Operations

```python
df1['Name'] = df1['Name'].str.title()         
df1['Department'] = df1['Department'].str.upper()  
```

üîó Step 6: Merging Data

```python
inner = pd.merge(df1, df2, on='Emp_ID', how='inner')
left = pd.merge(df1, df2, on='Emp_ID', how='left')
outer = pd.merge(df1, df2, on='Emp_ID', how='outer')
```

üìä Step 7: Data Operations ‚Äî Aggregation & GroupBy

```python
avg_salary = df1.groupby('Department')['Salary'].mean().reset_index()
print(avg_salary)
```

‚úÖ Step 8: Final Clean Data

```python
print(df1)
```

üìä Step 9: Visualization with Matplotlib

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
plt.bar(avg_salary['Department'], avg_salary['Salary'])
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Salary (‚Çπ)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
